```python
import pandas as pd
import numpy as np
```

```python
url_order_items = 'https://raw.githubusercontent.com/OtavioSotnas/Trabalhinho-Ada-PT2/main/base%20de%20dados/olist_order_items_dataset.csv'
```

```python
df_order_items = pd.read_csv(url_order_items)
```

```python
df_order_items = df_order_items.rename(columns={
    'order_id' : 'id_pedido',
    'order_item_id' : 'id_item_pedido',
    'product_id' : 'id_produto',
    'seller_id'  : 'id_vendedor',
    'shipping_limit_date' : 'data_limite_envio',
    'price' : 'valor',
    'freight_value' : 'valor_frete'
})
```

```python
df_order_items['valor_total'] = df_order_items['valor'] + df_order_items['valor_frete']
```

```python
df_order_items['ano_envio'] = df_order_items['data_limite_envio'].str.split('-').str[0]
```

```python
vendedor_mais_vendeu = df_order_items.groupby('id_vendedor')['valor'].sum().sort_values().tail(20)
```
